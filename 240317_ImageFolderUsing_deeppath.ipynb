```python
import os

# Define the path to the folder containing the files
svs_path = "/content/drive/MyDrive/Colab Notebooks/deepPatho/deeppath_utilizing_notebooks/follicular_neoplasm"

# Get a list of all files in the folder
files = os.listdir(svs_path)
count = 0
word = files[-1][-8:]
# Iterate through each file in the folder
for file in files:
    # print(file)
    # Check if the file name contains '의 사본'
    if word in str(file):
        print(file)
        # Extract the part of the file name before '의 사본'
        new_name = file.split(word)[0]

        # Construct the new file path
        old_path = os.path.join(svs_path, file)
        new_path = os.path.join(svs_path, new_name)
        print(count)
        count += 1

        # Rename the file
        os.rename(old_path, new_path)
```

```python
import os
import shutil

# Define the path to the folder containing the files
svs_path = "/content/drive/MyDrive/Colab Notebooks/deepPatho/deeppath_utilizing_notebooks/follicular_neoplasm"

# Get a list of all files in the folder
files = os.listdir(svs_path)

# Iterate through each file in the folder
for file in files:
    # Convert the file name to Unicode to handle encoding issues
    file = file.encode('utf-8').decode('utf-8')

    # Check if the file name contains '의 사본'
    if '의 사본' in file:
        # Extract the part of the file name before '의 사본'
        new_name = file.split('의 사본')[0]

        # Construct the old and new file paths
        old_path = os.path.join(svs_path, file)
        new_path = os.path.join(svs_path, new_name)

        # Rename the file
        shutil.move(old_path, new_path)
```

##타일링 실행_xml 파일과 연동하기

```python
#연동되는 xml 파일 확인
import re
import glob
# Define the paths
svs_path = "/content/drive/MyDrive/Colab Notebooks/deepPatho/deeppath_utilizing_notebooks/follicular_neoplasm"

# Get the list of SVS and XML files
svs_files = glob.glob(svs_path + "/*.svs")
xml_files = glob.glob(svs_path + "/*.xml")

# Function to extract file numbers from file paths
def extract_file_number(file_path):
    # Use regular expression to extract the file number
    match = re.search(r'(\d+)\.(svs|xml)$', file_path)
    if match:
        return match.group(1)
    else:
        return None

# Extract file numbers from SVS and XML files
svs_numbers = [extract_file_number(file) for file in svs_files]
xml_numbers = [extract_file_number(file) for file in xml_files]

# Filter out None values (in case of mismatched file names)
svs_numbers = [number for number in svs_numbers if number is not None]
xml_numbers = [number for number in xml_numbers if number is not None]

# Combine the lists and remove duplicates
file_numbers = xml_numbers
file_numbers.sort()

#이때부터 시작한다!
file_numbers = [number for number in file_numbers if int(number)>1028040]

# Print the file numbers
print(file_numbers)
```
